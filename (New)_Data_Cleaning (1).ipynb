```python
import pandas as pd
import numpy as np
import nltk
import string
import re
```

```python
df = pd.read_csv('TWITTER_MENTAH_BARU.csv', engine='python')
```

```python
df.head()
```

```python
df['tweet'] = df['tweet'].replace(np.nan, '')
```

Remove User

```python
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
    return input_txt    
df['remove_user'] = np.vectorize(remove_pattern)(df['tweet'], "@[\w]*")
```

```python
df.head()
```

```python
def remove(tweet):
    #remove angka
    tweet = re.sub('[0-9]+', '', tweet)
    
    # remove stock market tickers like $GE
    tweet = re.sub(r'\$\w*', '', tweet)
 
    # remove old style retweet text "RT"
    tweet = re.sub(r'^RT[\s]+', '', tweet)
    
    # remove hashtags
    # only removing the hashtag from the word
    tweet = re.sub(r'#[A-Za-z0-9]+', '', tweet)

    #remove https
    tweet = re.sub('(http|https):\/\/\S+', '',tweet)

    #remove punctuation
    tweet = re.sub(r'[^\w\s]', '', tweet)

    #lowercase
    tweet = tweet.lower()
    return tweet
df['remove_RT'] = df['remove_user'].apply(lambda x: remove(x))
```

```python
def remove_emojis(data):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', data)
df['remove_emoji'] = df['remove_RT'].apply(lambda x: remove_emojis(x))
df.head()
```
