```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
from nltk import pos_tag
from nltk.corpus import wordnet
nltk.download('averaged_perceptron_tagger')
nltk.download('omw-1.4')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cleaned_datasetbaru.csv')
df.info()
```

```python
#pos tagger dictionary
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}
```

```python
def token_stop_pos(text):
  tags = pos_tag(word_tokenize(text))
  newlist = []
  for word, tag in tags:
    if word.lower() not in set(stopwords.words('english')):
      newlist.append(tuple([word,pos_dict.get(tag[0])]))
  return newlist
    
```

```python
df['cleaned_tweets'] = df['cleaned_tweets'].replace(np.nan, '')
```

```python
df['Token,Stopwords, and POS tagged'] = df['cleaned_tweets'].apply(token_stop_pos)
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

def lemmatize(pos_data):
  lemma_r = " "
  for word, pos in pos_data:
    if not pos:
      lemma = word
      lemma_r = lemma_r + " " + lemma
    else:
      lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
      lemma_r = lemma_r + " " + lemma
  return lemma_r
df['Lemma'] = df['Token,Stopwords, and POS tagged'].apply(lemmatize)
df.head()
```
